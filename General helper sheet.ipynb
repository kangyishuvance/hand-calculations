```python
import math
from datetime import datetime, timedelta

def offset_by_weeks(date_str, weeks_offset):
    """
    date_str: 'YYYY-MM-DD'
    weeks_offset: integer (positive or negative)
    """
    base = datetime.strptime(date_str, "%Y-%m-%d")
    new_date = base + timedelta(weeks=weeks_offset)

    # ISO week number (ISO-8601)
    iso_year, iso_week, iso_weekday = new_date.isocalendar()

    print(f"Base date: {base.date()}")
    print(f"Offset (weeks): {weeks_offset}")
    print(f"New date: {new_date.date()}")
    print(f"ISO week: {iso_year}-W{iso_week}")

# Example
offset_by_weeks("2025-01-02", 5)
```

```python
from pathlib import Path

def find_zip_files(folder_path):
    folder = Path(folder_path)

    if not folder.exists():
        print("Folder does not exist")
        return

    zip_files = list(folder.rglob("*.zip"))

    if not zip_files:
        print("No zip files found")
        return

    for z in zip_files:
        print(z)

# usage
find_zip_files(r"C:\Users\vance\OneDrive\3) Work\4) BCA\ST Projects")
```
